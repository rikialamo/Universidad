```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("13_GameSales.csv")
```

Ahora vamos a generar un grafico que nos muestre cuantos juegos hay de cada plataforma

```python
plt.figure(figsize=(15, 10))
plt.pie(df['Platform'].value_counts())
plt.show()
```

```python
plt.savefig('grafica.png')
```
